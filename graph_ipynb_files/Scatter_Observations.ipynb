# **Code to Plot a Single Line Graph**

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Function to plot two line graphs
def plot_graph(x1, y1,matrix_size_1, x2, y2, matrix_size_2,x_axis_label, y_axis_label,title):
    plt.plot(x1, y1, marker='o',label=matrix_size_1)  # Plot the 128x128 line
    plt.plot(x2, y2, marker='o',label=matrix_size_2)  # Plot the 150x150 line
    x_min = min(min(x1), min(x2))          # Set limits on the x axis
    x_max = max(max(x1), max(x2))
    plt.xlim([x_min, x_max])

    y_min = min(min(y1), min(y2))          # Set limits on the y axis
    y_max = max(max(y1), max(y2))
    plt.ylim([y_min, y_max])

    plt.xlabel(x_axis_label)               # Label for the x-axis
    plt.ylabel(y_axis_label)               # Label for the y-axis
    plt.title(title)                       # Title of the graph
    plt.legend()                           # Add the legend
    plt.grid(True)                         # Enable grid
    plt.show()                             # Display the graph
```

# **Code to Plot Whole Graph Grid**

```python
# Function to plot the whole graph grid
def plot_graph_grid(data_list, grid_size=(3, 5)):
    fig, axes = plt.subplots(grid_size[0], grid_size[1], figsize=(40, 20))  # Adjust grid size and figure size

    # Flatten the axes array to easily loop through it (works even if it's 1x1)
    axes = axes.flatten()

    # Iterate over the data_list and plot the graphs in the corresponding subplot
    for i, (x1, y1, matrix_size_1, x2, y2, matrix_size_2 ,x_axis_label, y_axis_label) in enumerate(data_list):
        ax = axes[i]  # Select the correct subplot
        ax.plot(x1, y1, marker='o',markersize=15,linewidth=3,label=matrix_size_1)
        ax.plot(x2, y2, marker='o',markersize=15,linewidth=3,label=matrix_size_2)

        # Set limits for x and y axes
        x_min = min(min(x1), min(x2))
        x_max = max(max(x1), max(x2))
        y_min = min(min(y1), min(y2))
        y_max = max(max(y1), max(y2))
        ax.set_xlim([x_min, x_max])
        ax.set_ylim([y_min, y_max])

        # Set axis labels and title for each subplot
        ax.set_xlabel(x_axis_label,fontsize=30)
        ax.set_ylabel(y_axis_label,fontsize=30)
        ax.tick_params(axis='both', which='major', labelsize=25)
        ax.legend(fontsize=23)
        ax.yaxis.get_offset_text().set_fontsize(27)
        ax.grid(True)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

data_list = [
    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Depth
    ([1, 2, 3, 4], [50494314,7062586,3916690,3894010], '128x128', [1, 2, 3, 4], [57216435,6137707,4305903,4128911],'150x150','Depth','Clock Cycles'),
    ([1, 2, 3, 4], [37931,41693,28535,37931], '128x128', [1, 2, 3, 4], [43189,41461,34663,39215],'150x150','Depth','Misses'),
    ([1, 2, 3, 4], [23.22,25.52,17.47,23.22], '128x128',[1, 2, 3, 4], [25.79,24.76,20.70,23.42],'150x150','Depth','Miss Rate(%)'),
    ([1, 2, 3, 4], [76.77,74.47,82.52,76.77], '128x128',[1, 2, 3, 4], [74.20,75.23,79.29,76.57],'150x150','Depth','Hit Rate(%)'),
    ([1, 2, 3, 4], [233.25,32.76,16.99,17.91], '128x128',[1, 2, 3, 4], [258.94,29.81,17.87,16.73],'150x150','Depth','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Block Size
    ([16,32,64,128], [4276086,3555318,3587126,3085046], '128x128', [16,32,64,128], [4373335,3783927,3499799,3369131], '150x150','Block Size','Clock Cycles'),
    ([16,32,64,128], [41693,10737,15151,4259], '128x128', [16,32,64,128], [37869,12077,8596,5094], '150x150','Block Size','Misses'),
    ([16,32,64,128], [25.52,6.57,9.27,2.60], '128x128', [16,32,64,128], [22.61,7.21,5.13,3.04], '150x150','Block Size','Miss Rate(%)'),
    ([16,32,64,128], [74.47,93.42,90.72,97.39], '128x128', [16,32,64,128], [77.38,92.78,94.86,96.95], '150x150','Block Size','Hit Rate(%)'),
    ([16,32,64,128], [18.30,16.14,15.74,16.23], '128x128', [16,32,64,128], [17.56,17.72,17.20,17.41], '150x150','Block Size','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Associativity
    ([2,4,8], [4087938,4151650,3749362] , '128x128', [2,4,8], [4342911,4090515,4265699], '150x150','Associativity','Clock Cycles'),
    ([2,4,8], [34301,41510,23860], '128x128', [2,4,8], [39215,32275,38855], '150x150','Associativity','Misses'),
    ([2,4,8], [21.00,25.41,14.60], '128x128', [2,4,8], [23.42,19.27,23.20], '150x150','Associativity','Miss Rate(%)'),
    ([2,4,8], [78.99,74.58,85.39], '128x128', [2,4,8], [76.57,80.72,76.79], '150x150','Associativity','Hit Rate(%)'),
    ([2,4,8], [17.00,19.79,17.37], '128x128', [2,4,8], [17.75,19.51,20.83], '150x150','Associativity','AMAT'),


]

# Plot the whole grid
plot_graph_grid(data_list, grid_size=(3, 5))
```
