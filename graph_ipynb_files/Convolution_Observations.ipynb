# **Code to Plot a Single Line Graph**

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Function to plot two line graphs
def plot_graph(x1, y1,matrix_size_1, x2, y2, matrix_size_2,x_axis_label, y_axis_label,title):
    plt.plot(x1, y1, marker='o',label=matrix_size_1)  # Plot the 32x32 line
    plt.plot(x2, y2, marker='o',label=matrix_size_2)  # Plot the 20x20 line
    x_min = min(min(x1), min(x2))          # Set limits on the x axis
    x_max = max(max(x1), max(x2))
    plt.xlim([x_min, x_max])

    y_min = min(min(y1), min(y2))          # Set limits on the y axis
    y_max = max(max(y1), max(y2))
    plt.ylim([y_min, y_max])

    plt.xlabel(x_axis_label)               # Label for the x-axis
    plt.ylabel(y_axis_label)               # Label for the y-axis
    plt.title(title)                       # Title of the graph
    plt.legend()                           # Add the legend
    plt.grid(True)                         # Enable grid
    plt.show()                             # Display the graph
```

# **Code to Plot Whole Graph Grid**

```python
# Function to plot the whole graph grid
def plot_graph_grid(data_list, grid_size=(3, 5)):
    fig, axes = plt.subplots(grid_size[0], grid_size[1], figsize=(40, 20))  # Adjust grid size and figure size

    # Flatten the axes array to easily loop through it (works even if it's 1x1)
    axes = axes.flatten()

    # Iterate over the data_list and plot the graphs in the corresponding subplot
    for i, (x1, y1, matrix_size_1, x2, y2, matrix_size_2 ,x_axis_label, y_axis_label) in enumerate(data_list):
        ax = axes[i]  # Select the correct subplot
        ax.plot(x1, y1, marker='o',markersize=15,linewidth=3,label=matrix_size_1)
        ax.plot(x2, y2, marker='o',markersize=15,linewidth=3,label=matrix_size_2)

        # Set limits for x and y axes
        x_min = min(min(x1), min(x2))
        x_max = max(max(x1), max(x2))
        y_min = min(min(y1), min(y2))
        y_max = max(max(y1), max(y2))
        ax.set_xlim([x_min, x_max])
        ax.set_ylim([y_min, y_max])

        # Set axis labels and title for each subplot
        ax.set_xlabel(x_axis_label,fontsize=30)
        ax.set_ylabel(y_axis_label,fontsize=30)
        ax.tick_params(axis='both', which='major', labelsize=25)
        ax.legend(fontsize=23)
        ax.yaxis.get_offset_text().set_fontsize(27)
        ax.grid(True)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

data_list = [
    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Depth
    ([1, 2, 3, 4], [87702423,10654487,5897423,5316055], '32x32', [1, 2, 3, 4], [34596193,5284377,2931977,2499613],'20x20','Depth','Clock Cycles'),
    ([1, 2, 3, 4], [72453,72453,75546,62363], '32x32', [1, 2, 3, 4], [28685,26668,26668,30973],'20x20','Depth','Misses'),
    ([1, 2, 3, 4], [10.37,10.37,10.81,8.92], '32x32',[1, 2, 3, 4], [10.12,9.41,9.41,10.93],'20x20','Depth','Miss Rate(%)'),
    ([1, 2, 3, 4], [89.62,89.62,89.18,91.07], '32x32',[1, 2, 3, 4], [89.87,90.58,90.58,89.06],'20x20','Depth','Hit Rate(%)'),
    ([1, 2, 3, 4], [104.73,12.70,6.64,5.965], '32x32',[1, 2, 3, 4], [102.29,14.77,8.21,7.54],'20x20','Depth','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Block Size
    ([16,32,64,128], [5966563,4740107,4209807,3931243], '32x32', [16,32,64,128], [2961501,2645125,2350021,2114825], '20x20','Block Size','Clock Cycles'),
    ([16,32,64,128], [78851,23833,10253,6881], '32x32', [16,32,64,128], [29073,12357,5933,2722], '20x20','Block Size','Misses'),
    ([16,32,64,128], [11.28,3.41,1.46,0.98], '32x32', [16,32,64,128], [10.26,4.36,2.09,0.96], '20x20','Block Size','Miss Rate(%)'),
    ([16,32,64,128], [88.71,96.58,98.53,99.01], '32x32', [16,32,64,128], [89.73,95.63,97.90,99.03], '20x20','Block Size','Hit Rate(%)'),
    ([16,32,64,128], [6.74,5.15,4.73,4.64], '32x32', [16,32,64,128], [8.23,7.31,6.71,6.55], '20x20','Block Size','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Associativity
    ([2,4,8], [5897423,5211803,5030555] , '32x32', [2,4,8], [2957501,2790293,2765581], '20x20','Associativity','Clock Cycles'),
    ([2,4,8], [75546,49455,41716], '32x32', [2,4,8], [29073,21206,19151], '20x20','Associativity','Misses'),
    ([2,4,8], [10.81,7.08,5.97], '32x32', [2,4,8], [10.26,7.48,6.76], '20x20','Associativity','Miss Rate(%)'),
    ([2,4,8], [89.18,92.91,94.02], '32x32', [2,4,8], [89.73,92.51,93.23], '20x20','Associativity','Hit Rate(%)'),
    ([2,4,8], [6.64,5.38,5.22], '32x32', [2,4,8], [8.21,7.30,7.26], '20x20','Associativity','AMAT'),


]

# Plot the whole grid
plot_graph_grid(data_list, grid_size=(3, 5))
```
