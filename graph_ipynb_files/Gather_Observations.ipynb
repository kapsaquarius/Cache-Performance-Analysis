# **Code to Plot a Single Line Graph**

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Function to plot two line graphs
def plot_graph(x1, y1,matrix_size_1, x2, y2, matrix_size_2,x_axis_label, y_axis_label,title):
    plt.plot(x1, y1, marker='o',label=matrix_size_1)  # Plot the 128x128 line
    plt.plot(x2, y2, marker='o',label=matrix_size_2)  # Plot the 150x150 line
    x_min = min(min(x1), min(x2))          # Set limits on the x axis
    x_max = max(max(x1), max(x2))
    plt.xlim([x_min, x_max])

    y_min = min(min(y1), min(y2))          # Set limits on the y axis
    y_max = max(max(y1), max(y2))
    plt.ylim([y_min, y_max])

    plt.xlabel(x_axis_label)               # Label for the x-axis
    plt.ylabel(y_axis_label)               # Label for the y-axis
    plt.title(title)                       # Title of the graph
    plt.legend()                           # Add the legend
    plt.grid(True)                         # Enable grid
    plt.show()                             # Display the graph
```

# **Code to Plot Whole Graph Grid**

```python
# Function to plot the whole graph grid
def plot_graph_grid(data_list, grid_size=(3, 5)):
    fig, axes = plt.subplots(grid_size[0], grid_size[1], figsize=(40, 20))  # Adjust grid size and figure size

    # Flatten the axes array to easily loop through it (works even if it's 1x1)
    axes = axes.flatten()

    # Iterate over the data_list and plot the graphs in the corresponding subplot
    for i, (x1, y1, matrix_size_1, x2, y2, matrix_size_2 ,x_axis_label, y_axis_label) in enumerate(data_list):
        ax = axes[i]  # Select the correct subplot
        ax.plot(x1, y1, marker='o',markersize=15,linewidth=3,label=matrix_size_1)
        ax.plot(x2, y2, marker='o',markersize=15,linewidth=3,label=matrix_size_2)

        # Set limits for x and y axes
        x_min = min(min(x1), min(x2))
        x_max = max(max(x1), max(x2))
        y_min = min(min(y1), min(y2))
        y_max = max(max(y1), max(y2))
        ax.set_xlim([x_min, x_max])
        ax.set_ylim([y_min, y_max])

        # Set axis labels and title for each subplot
        ax.set_xlabel(x_axis_label,fontsize=30)
        ax.set_ylabel(y_axis_label,fontsize=30)
        ax.tick_params(axis='both', which='major', labelsize=25)
        ax.legend(fontsize=23)
        ax.yaxis.get_offset_text().set_fontsize(27)
        ax.grid(True)

    # Adjust layout to prevent overlap
    plt.tight_layout()
    plt.show()

data_list = [
    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Depth
    ([1, 2, 3, 4], [87200883,25387107,13616395,13354795], '128x128', [1, 2, 3, 4], [91687240,28185824,16042196,16134596],'150x150','Depth','Clock Cycles'),
    ([1, 2, 3, 4], [70907,69462,74816,68727], '128x128', [1, 2, 3, 4], [74471,77597,77916,77916],'150x150','Depth','Misses'),
    ([1, 2, 3, 4], [24.63,24.12,25.98,23.87], '128x128',[1, 2, 3, 4], [24.08,25.09,25.19,25.19],'150x150','Depth','Miss Rate(%)'),
    ([1, 2, 3, 4], [75.36,75.87,74.01,76.12], '128x128',[1, 2, 3, 4], [75.91,74.90,74.80,74.80],'150x150','Depth','Hit Rate(%)'),
    ([1, 2, 3, 4], [247.30,73.04,41.06,39.49], '128x128',[1, 2, 3, 4], [241.81,74.77,45.49,44.27],'150x150','Depth','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Block Size
    ([16,32,64,128], [13509795,12829595,12006335,11234647], '128x128', [16,32,64,128], [15995324,15576960,14377620,13788116], '150x150','Block Size','Clock Cycles'),
    ([16,32,64,128], [70758,38638,20544,13258], '128x128', [16,32,64,128], [77597,47770,26606,17668], '150x150','Block Size','Misses'),
    ([16,32,64,128], [24.57,13.42,7.13,4.60], '128x128', [16,32,64,128], [25.09,15.44,8.60,5.71], '150x150','Block Size','Miss Rate(%)'),
    ([16,32,64,128], [75.42,86.57,92.86,95.39], '128x128', [16,32,64,128], [74.90,84.55,91.39,94.28], '150x150','Block Size','Hit Rate(%)'),
    ([16,32,64,128], [40.15,38.75,37.15,35.37], '128x128', [16,32,64,128], [43.75,41.99,42.14,39.61], '150x150','Block Size','AMAT'),

    # (Clock Cycles, Misses, Miss Rate(%), Hit Rate(%), AMAT) vs Associativity
    ([2,4,8], [13472591,13539219,13420251] , '128x128', [2,4,8], [16089412,16074720,15925388], '150x150','Associativity','Clock Cycles'),
    ([2,4,8], [70907,71450,72132], '128x128', [2,4,8], [76209,78529,70046], '150x150','Associativity','Misses'),
    ([2,4,8], [24.63,24.81,25.05], '128x128', [2,4,8], [24.64,25.39,22.65], '150x150','Associativity','Miss Rate(%)'),
    ([2,4,8], [75.36,75.18,74.94], '128x128', [2,4,8], [75.35,74.60,77.34], '150x150','Associativity','Hit Rate(%)'),
    ([2,4,8], [41.48,39.88,40.04], '128x128', [2,4,8], [45.87,44.17,43.36], '150x150','Associativity','AMAT'),


]

# Plot the whole grid
plot_graph_grid(data_list, grid_size=(3, 5))
```
